```python
import pandas as pd
```

```python
in_df = pd.read_csv("../Trails For Project/ProjDataset/archive/IN_youtube_trending_data.csv")
in_df.head(2)
```

```python
in_df["year_month"] = [i[0:7] for i in in_df["trending_date"]]
```

```python
in_df.head(1)
```

```python
for key,df in in_df.groupby(['year_month']):
    filename = "./Data/IN/IN_"+key+".csv"
    df_new = df.iloc[:,:-2].reset_index(drop=True)
    print(df_new.head(2))
    df_new.to_csv(filename)
    print(df.iloc[:,:-2].shape)
```
